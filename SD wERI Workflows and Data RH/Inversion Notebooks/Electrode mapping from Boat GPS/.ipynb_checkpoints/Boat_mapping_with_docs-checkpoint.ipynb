# Map filtered boat GPS to individual surveys
---
Assumes the clock from the resistivity meter and the clock from the boat GPS controller were syncronized. 
This will match X and Y pairs from boat_GPS to the ERT data frame. The (X,Y) pair represent the location of the boat at a specific time. The Easting and Northing of the first electrodes was defined to be 2 meters behind the boat GPS point along the same bearing and all other 12 electrodes have a 1.5 m. 

The boat GPS pair (x1, y2) and up to 6 meters of the consecutive points trailing x1 and x2 were used in a linear regression.

$\theta_{survey} = tan^{-1}(m_r)$

$lx_{el} = (l - dist)*cos(\theta_{survey})$

$ly_{el} = (l - dist)*sin(\theta_{survey})$

where l is the length of the regression line (~6 meters) and dist is the distance to a specific electrode along a line.

# Points that lie along the regression line are calculated at 
---

el1 = 2m, el2 = 3.5, el3 = 5.0, el4 = 6.5, el5 = 8.0, 

el6 = 9.5, el7 = 11, el8 = 12.5, el9 = 14.0, el10 = 15.5

el11 = 17.0, el12 = 18.5, el13 = 20.0

```python
def match_ert_to_boat(
    ert, 
    boat_gps, 
    time_threshold=5, 
    time_col_ert="Date", 
    time_col_gps="Time", 
    x_col="x", 
    y_col="y"
):
    """
    Match ERT times to boat GPS times with a maximum threshold in seconds.
    Ensures unique matches (each boat_gps row used at most once).

    Parameters
    ----------
    ert : pd.DataFrame
        Dataframe containing ERT data with datetime column.
    boat_gps : pd.DataFrame
        Dataframe containing GPS data with time column + x, y coordinates.
    time_threshold : int
        Max allowed difference in seconds between ERT and GPS times.
    time_col_ert : str
        Column name of datetime in ert (format like "1/28/2024 5:24:51 PM").
    time_col_gps : str
        Column name of time in boat_gps (format like "17:24:01").
    x_col, y_col : str
        Column names for boat GPS coordinates.

    Returns
    -------
    ert_sorted : pd.DataFrame
        ert dataframe with added 'x_mapped', 'y_mapped', and 'Matched' columns.
    boat_sorted : pd.DataFrame
        boat_gps dataframe with an added 'Matched' column (1 if used, 0 otherwise).
    """

    # --- Parse ERT full datetime with AM/PM ---
    ert_parsed = pd.to_datetime(
        ert[time_col_ert], 
        format="%m/%d/%Y %I:%M:%S %p",  
        errors="coerce"
    )

    # --- Parse boat GPS times (24-hour format) ---
    gps_parsed = pd.to_datetime(
        boat_gps[time_col_gps], 
        format="%H:%M:%S",  
        errors="coerce"
    ).dt.time  

    # --- Put both onto the same dummy date ---
    dummy_date = pd.to_datetime("2000-01-01")
    ert_times = pd.to_datetime(ert_parsed.dt.time.astype(str), format="%H:%M:%S") + pd.Timedelta(days=dummy_date.day-1)
    gps_times = pd.to_datetime(gps_parsed.astype(str), format="%H:%M:%S") + pd.Timedelta(days=dummy_date.day-1)

    # --- Sort both dataframes by time ---
    ert_sorted = ert.copy()
    ert_sorted["Time_fixed"] = ert_times
    ert_sorted = ert_sorted.sort_values("Time_fixed").reset_index(drop=True)
    ert_sorted["x_mapped"] = np.nan
    ert_sorted["y_mapped"] = np.nan
    ert_sorted["Matched"] = 0  # initialize as unmatched

    boat_sorted = boat_gps.copy()
    boat_sorted["Time_fixed"] = gps_times
    boat_sorted = boat_sorted.sort_values("Time_fixed").reset_index(drop=True)
    boat_sorted["Matched"] = 0  

    used_gps_idx = set()

    # --- Match loop ---
    for i, ert_time in enumerate(ert_sorted["Time_fixed"]):
        diffs = (boat_sorted["Time_fixed"] - ert_time).dt.total_seconds().abs()
        if diffs.empty:
            continue

        nearest_idx = diffs.idxmin()
        min_diff = diffs.loc[nearest_idx]  

        # Only accept if within threshold and GPS not already used
        if min_diff <= time_threshold and nearest_idx not in used_gps_idx:
            ert_sorted.loc[i, "x_mapped"] = boat_sorted.loc[nearest_idx, x_col]
            ert_sorted.loc[i, "y_mapped"] = boat_sorted.loc[nearest_idx, y_col]
            ert_sorted.loc[i, "Matched"] = 1  # mark ERT row as matched
            used_gps_idx.add(nearest_idx)
            boat_sorted.loc[nearest_idx, "Matched"] = 1  

    return ert_sorted, boat_sorted
    
# ---------------- Example usage ----------------
ert_with_gps, boat_gps_with_flags = match_ert_to_boat(
    ert, boat_gps,
    time_threshold=2,
    time_col_ert="Date", time_col_gps="Time",
    x_col="X", y_col="Y"
)
```

```python
def filter_ert_by_matched_blocks(ert_with_gps, block_size=10):
    """
    Keep rows in ert_with_gps where Matched == 1 and the next (block_size - 1) rows.
    Removes other rows.
    
    Parameters
    ----------
    ert_with_gps : pd.DataFrame
        DataFrame containing 'Matched' column (1 if matched to GPS, 0 otherwise).
    block_size : int
        Number of consecutive rows to keep starting from a matched row.
        
    Returns
    -------
    pd.DataFrame
        Filtered ERT rows.
    """
    keep_indices = []

    matched_indices = ert_with_gps.index[ert_with_gps['Matched'] == 1]

    for idx in matched_indices:
        # Add idx and the next (block_size - 1) rows
        block_indices = list(range(idx, min(idx + block_size, len(ert_with_gps))))
        keep_indices.extend(block_indices)

    # Remove duplicates and sort
    keep_indices = sorted(set(keep_indices))

    return ert_with_gps.loc[keep_indices].reset_index(drop=True)


# --- Example usage ---
ert_filtered = filter_ert_by_matched_blocks(ert_with_gps, block_size=10)

print(f"Original length: {len(ert_filtered)}, Filtered length: {len(boat_gps_with_flags)}")
# print(f"Should be roughly {len(boat_gps) * 10} rows: {len(ert_filtered)}")
```

```python
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def segment_gps_tracks(boat_gps_with_flags, min_dist=18, max_dist=19):
    """
    Segments GPS track backwards from rows where 'Matched' == 1
    into chunks of 18â€“19 meters, preserving all columns.
    Adds a segment_id column for each segment.
    """
    segments = []
    segment_counter = 0

    matched_indices = boat_gps_with_flags.index[boat_gps_with_flags['Matched'] == 1]

    for idx in matched_indices:
        cum_dist = 0.0
        segment_rows = [idx]

        for prev_idx in range(idx - 1, -1, -1):
            x1, y1 = boat_gps_with_flags.loc[prev_idx + 1, ['X', 'Y']]
            x2, y2 = boat_gps_with_flags.loc[prev_idx, ['X', 'Y']]

            step_dist = euclidean_distance(x1, y1, x2, y2)
            cum_dist += step_dist

            segment_rows.append(prev_idx)

            if cum_dist >= min_dist:
                if cum_dist <= max_dist:
                    segment = boat_gps_with_flags.loc[sorted(segment_rows)].copy()
                    segment["segment_id"] = segment_counter  # assign id
                    segments.append(segment)
                    segment_counter += 1
                break
    return segments

# Example usage
segments = segment_gps_tracks(boat_gps_with_flags, min_dist=0.5, max_dist=3)
```

```python
def compute_segment_regressions(segments, r2_threshold=0.6):
    """
    Compute linear regression for each segment, iteratively removing last
    points to improve RÂ² to >= r2_threshold. Stores summary info in a DataFrame.

    Returns
    -------
    final_segments : list of pd.DataFrame
        Segments after trimming points to improve RÂ²
    original_segments : list of pd.DataFrame
        Original untrimmed segments
    regression_summary : pd.DataFrame
        Regression summary for each final segment, including total distance, Pearson r, and angle
    """
    results = []
    final_segments = []

    for i, original_seg in enumerate(segments):
        seg = original_seg.copy()
        original_n_points = len(seg)
        points_removed = 0
        best_r2 = -np.inf
        best_model_stats = {'slope': np.nan, 'intercept': np.nan, 'r2': np.nan}
        best_n_points = len(seg)
        final_distance = np.nan
        final_pearson_r = np.nan
        theta_deg = np.nan

        while len(seg) >= 2:
            X = seg[['X']].values
            y = seg['Y'].values
            model = LinearRegression().fit(X, y)
            r2 = model.score(X, y)
            
            # Pearson correlation coefficient
            pearson_r, _ = pearsonr(seg['X'], seg['Y'])

            if r2 >= r2_threshold:
                best_r2 = r2
                best_model_stats = {
                    'slope': model.coef_[0],
                    'intercept': model.intercept_,
                    'r2': r2
                }
                best_n_points = len(seg)
                final_pearson_r = pearson_r
                theta_deg = np.degrees(np.arctan(model.coef_[0]))  # Angle of regression line
                # Compute total distance along segment
                dx = np.diff(seg['X'].values)
                dy = np.diff(seg['Y'].values)
                final_distance = np.sum(np.sqrt(dx**2 + dy**2))
                break  # satisfied threshold
            else:
                # remove last point
                seg = seg.iloc[:-1]
                points_removed += 1

        # Append final segment
        final_segments.append(seg)

        # Compute distance, Pearson r, and angle even if RÂ² never reached threshold
        if np.isnan(final_distance) and len(seg) >= 2:
            dx = np.diff(seg['X'].values)
            dy = np.diff(seg['Y'].values)
            final_distance = np.sum(np.sqrt(dx**2 + dy**2))
            final_pearson_r, _ = pearsonr(seg['X'], seg['Y'])
            theta_deg = np.degrees(np.arctan(best_model_stats['slope']))

        # Store regression stats
        results.append({
            'segment_id': i,
            'original_points': original_n_points,
            'points_removed': points_removed,
            'slope': best_model_stats['slope'],
            'intercept': best_model_stats['intercept'],
            'r2': best_model_stats['r2'],
            'pearson_r': final_pearson_r,
            'theta_deg': theta_deg,
            'remaining_points': best_n_points,
            'total_distance': final_distance
        })

        # Print info for this segment
        if best_r2 >= r2_threshold:
            print(f"Segment {i}: Original points={original_n_points}, Points removed={points_removed}, "
                  f"Final RÂ²={best_r2:.3f}, Pearson r={final_pearson_r:.3f}, "
                  f"Angle={theta_deg:.2f}Â°, Remaining points={best_n_points}, Total distance={final_distance:.2f}")
        else:
            print(f"Segment {i}: Original points={original_n_points}, Points removed={points_removed}, "
                  f"RÂ² never reached {r2_threshold}, Final RÂ²={best_r2:.3f}, Pearson r={final_pearson_r:.3f}, "
                  f"Angle={theta_deg:.2f}Â°, Remaining points={len(seg)}, Total distance={final_distance:.2f}")

    return final_segments, segments, pd.DataFrame(results)

def generate_points_along_regression_gps(final_segments, regression_summary, n_points=13, first_offset=2, spacing=1.5):
    """
    Generate fixed-offset points along the regression line for each segment,
    in the original GPS coordinates.

    Parameters
    ----------
    final_segments : list of pd.DataFrame
        The segmented GPS data.
    regression_summary : pd.DataFrame
        Regression stats including slope, theta_deg.
    n_points : int
        Number of points to generate per segment.
    first_offset : float
        Distance of the first electrode from the starting GPS point (meters).
    spacing : float
        Spacing between subsequent electrodes (meters).

    Returns
    -------
    points_list : list of pd.DataFrame
        Each DataFrame has n_points rows of GPS X,Y.
    """
    points_list = []

    for i, seg in enumerate(final_segments):
        row = regression_summary.loc[regression_summary['segment_id'] == i].iloc[0]
        theta = np.radians(row['theta_deg'])

        # First point of segment in GPS coords
        x0, y0 = seg['X'].iloc[0], seg['Y'].iloc[0]

        # Distances along the regression line: first_offset + spacing*(k-1)
        step_distances = first_offset + np.arange(n_points) * spacing

        # Convert to X, Y GPS coordinates
        x = x0 + step_distances * np.cos(theta)
        y = y0 + step_distances * np.sin(theta)

        points_df = pd.DataFrame({'X': x, 'Y': y})
        points_list.append(points_df)

    return points_list

def plot_segments_with_r2(final_segments, original_segments, regression_summary): 

    r2_vals = regression_summary['r2'].fillna(0).values
    norm = plt.Normalize(vmin=min(r2_vals), vmax=max(r2_vals))
    cmap = cm.viridis
    
    plt.figure(figsize=(8, 6))
    
    for i, final_seg in enumerate(final_segments):
        original_seg = original_segments[i]
        
        r2 = regression_summary.loc[regression_summary['segment_id'] == i, 'r2'].values[0]
        
        # Get color for the main segment based on RÂ² value
        color = cmap(norm(r2 if not np.isnan(r2) else 0))
        if r2 > 0.6:
            color = 'k'
        
        # Plot the final, trimmed segment points
        plt.scatter(
            final_seg['X'], 
            final_seg['Y'], 
            c=[color] * len(final_seg), 
            alpha=0.7, 
            label=f'Segment {i}' if i == 0 else ""
        )
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("GPS Segments")
    plt.tight_layout()
    plt.show()

# Example usage
final_segments, original_segments, regression_summary = compute_segment_regressions(segments)
points_list_gps = generate_points_along_regression_gps(final_segments, regression_summary)
plot_segments_with_r2(final_segments, original_segments, regression_summary)
```

```python
# Initialize list to hold all rows
# all_rows = []

# for seg_idx, df in enumerate(points_list_gps, start=1):
#     for row_idx, row in df.iterrows():
#         label = f"{seg_idx} {row_idx + 1}"
#         x, y, z = row['X'], row['Y'], 0
#         buried = 0 if z == 0 else 1
#         all_rows.append({'label': label, 'x': x, 'y': y, 'z': z, 'buried': buried})

# Create single combined DataFrame
# combined_df = pd.DataFrame(all_rows)

# Export to CSV
# combined_df.to_csv(r"C:\Users\jht51\Box\AFRI (3 Projects)\Waterborne Data\wERT_elec_combined.csv", index=False)
# ert_with_gps.to_csv(r"C:\Users\jht51\Box\AFRI (3 Projects)\Waterborne Data\wERT_combined.csv", index=False)
# print(f"Exported {len(combined_df)/13} rows to combined_points.csv")
# --- Original code snippet ends here ---

# Your request: Change the code to export a new CSV file for every 13 rows
# from the 'points_list_gps' data.

# Here is the modified code that achieves this.

def export_segments_by_rows(points_list_gps, rows_per_file=13, base_path=r"C:\Users\jht51\Box\AFRI (3 Projects)\Papers for P Loss\ERT-based calibration of EMI\Waterborne\Electrode mapping from Boat GPS\ERT Electrode Files\Elec"):
    """
    Exports a new CSV file for every 'rows_per_file' rows of data
    from the points_list_gps DataFrame.

    Parameters
    ----------
    points_list_gps : list of pd.DataFrame
        List of DataFrames, where each DataFrame is a segment of GPS points.
    rows_per_file : int, optional
        The number of rows to include in each output CSV file. Defaults to 13.
    base_path : str, optional
        The base directory to save the output CSV files.
    """
    row_count = 0
    file_num = 1
    current_rows = []

   
    # Iterate through each segment (DataFrame) in the list
    for seg_idx, df in enumerate(points_list_gps, start=1):
        for row_idx, row in df.iterrows():
            # label = f"{seg_idx} {row_idx + 1}"
            # label = np.arange(1, 14, 1)
            x, y, z = row['X'], row['Y'], 0
            buried = 0 if z == 0 else 1
            # current_rows.append({'label': label, 'x': x, 'y': y, 'z': z, 'buried': buried})
            current_rows.append({'x': x, 'y': y, 'z': z})

            row_count += 1

            # Check if we have collected enough rows for a new file
            if row_count == rows_per_file:
                # Create a DataFrame from the collected rows
                segment_df = pd.DataFrame(current_rows)
                
                # Construct the filename and export to CSV
                output_filename = f"wERT_elec_segment_{file_num:03d}.csv"
                output_filepath = base_path + output_filename
                
                segment_df.to_csv(output_filepath, index=False)
                
                print(f"Exported {row_count} rows to {output_filepath}")

                # Reset for the next file
                current_rows = []
                row_count = 0
                file_num += 1

    # Handle any remaining rows that don't make a full segment of 13
    if current_rows:
        segment_df = pd.DataFrame(current_rows)
        output_filename = f"wERT_elec_segment_{file_num:03d}.csv"
        output_filepath = base_path + output_filename
        
        segment_df.to_csv(output_filepath, index=False)
        
        print(f"Exported last {len(current_rows)} rows to {output_filepath}")

export_segments_by_rows(points_list_gps)
```
